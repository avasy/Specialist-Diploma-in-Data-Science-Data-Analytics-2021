```python
import matplotlib.pyplot as plt #import this first
fig1 = plt.figure(figsize=(10,5))
fig1.set_facecolor("beige")
ax1 = fig1.add_subplot(121)
#121 -> 1 row, 2 columns, first cell
ax1.plot([0,1,2,3],[55,90,65,88])
ax1.set_facecolor("green")
ax2 = fig1.add_subplot(122)
ax2.plot([55,66,77,33])
ax2.set_facecolor("pink")
plt.show()
```

```python
import matplotlib.pyplot as plt #import this first
fig1 = plt.figure(figsize=(10,5))
fig1.set_facecolor("beige")
ax1 = fig1.add_subplot(121)
#121 -> 1 row, 2 columns, first cell
ax1.plot([0,1,2,3],[55,90,65,88])
ax1.set_facecolor("green")
ax2 = fig1.add_subplot(122)
ax2.plot([55,66,77,33])
ax2.set_facecolor("pink")
fig1.suptitle("This is the main title!", fontsize=14, fontweight='bold')
ax1.set_title("Ax1's title")
ax2.set_title("Ax2's title")
ax1.set_xticks([0,1,2,3]) #set location 
ax1.set_xticklabels(["a","b","c","d"]) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12, 8))  #width 12, height 8
data =[55,90,65,88]
names = ["Jess", "Jenny", "Joey", "John"]

plt.xticks(np.arange(len(data)), names)      
plt.bar(np.arange(len(data)),data)
plt.ylim(0,100)
x = 0
for y in data:
    plt.text(x-0.05,y,str(y), fontsize=15)
    x = x + 1

plt.show()
```
