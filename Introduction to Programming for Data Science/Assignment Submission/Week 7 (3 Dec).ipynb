```python
import numpy as np
import matplotlib.pyplot as plt

#correlation between weather temperature and ice cream sales
temp = np.random.randint(10, 40, 100)  #100 random numbers in the range of 10 to 39 degree
sales = np.random.randint(100, 501, 100)  #100 random numbers in the range of 100 to 500
plt.xlabel("Average Daily Temperature")
plt.ylabel("Ice Cream Sales")
plt.scatter(temp, sales)
m,b = np.polyfit(temp,sales,1) #regression line
plt.plot(temp, m * temp + b, "r--" ) #y = mx + b
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = "icecream.csv"

d = np.genfromtxt(f, delimiter=',', skip_header=True,
                  dtype=[('temp', 'f8'),('sales', 'i4')])
x = d['temp']
y= d['sales']
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Sales")
plt.scatter(x,y)
m,b = np.polyfit(x,y, 1)
plt.plot(x, m*x+b, "r--")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#correlation between weather temperature and ice cream sales
temp = np.random.randint(10, 40, 100)  #100 random numbers in the range of 10 to 39 degree
sales = np.random.randint(100, 501, 100)  #100 random numbers in the range of 100 to 500
plt.xlabel("Average Daily Temperature")
plt.ylabel("Ice Cream Sales")
colors = np.random.randint(1,100,len(temp )) #random no between 1 and 99
sizes = 10* np.random.randint(1,100,len(temp )) #random no between 1 and 99
plt.scatter(temp, sales,c = colors, s = sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
m,b = np.polyfit(temp,sales,1)
plt.plot(temp, m * temp + b, "r--" ) #y = mx + b
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = "median-resale-prices-for-registered-applications-by-town-and-flat-type.csv"

d = np.genfromtxt(f, delimiter=',', skip_header=True,
                  dtype=[('qtr', 'U7'),('town', 'U30'),('type','U20'),('price','i4')])
d = d[d['price'] != -1] #remove rows with missing prices
d = d[(d['type'] == '4-room') | (d['type'] == '4-ROOM')  ] #take only 4-room data
c = d[(d['town'] == 'Clementi') | (d['town'] == 'CLEMENTI')  ]
b = d[(d['town'] == 'Bishan') | (d['town'] == 'BISHAN')  ]
years = np.arange(2010, 2020)  #2010 to 2019
avgC, avgB = [], []
for yr in years: #for 2010 to 2019
    tmp = c[np.char.find(c['qtr'], str(yr)) > -1]
    avgPrice = tmp['price'].mean()
    avgC.append(avgPrice)
    tmp = b[np.char.find(b['qtr'], str(yr)) > -1]
    avgPrice = tmp['price'].mean()
    avgB.append(avgPrice)
loc = np.arange(len(years))
plt.xticks(loc, years, rotation=45)
plt.plot(avgC,label="Clementi")
plt.plot(avgB, label="Bishan")
plt.legend()
plt.show()
```
