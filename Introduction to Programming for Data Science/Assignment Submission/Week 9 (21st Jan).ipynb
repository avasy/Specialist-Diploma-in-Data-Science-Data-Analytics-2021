```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("orders.xlsx")
r = pd.unique(df['Region'])
totalSales = []
for x in r:
    tmp = df[df['Region'] == x]
    avg = tmp['Total'].sum()
    totalSales.append(avg)

plt.xticks(np.arange(len(r)), r)
plt.bar(np.arange(len(r)), totalSales)
```

```python
import pandas as pd
df = pd.read_excel("orders.xlsx")

#create a groupby object
g = df.groupby('Region')
x = g.sum()[['Total']] #x = dataframe containing total sales of each region
x['Total'].plot(kind='bar')
```
